```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
```

### The Curve
Let's start by examining how coronavirus deaths have increased over time.

```python
deaths_time = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_US.csv')
dates = list(deaths_time.columns[12:])
deaths_by_date = deaths_time[dates].sum(axis=0)
deaths_by_date.index = pd.to_datetime(deaths_by_date.index)

chart = sns.lineplot(
    x = deaths_by_date.index, 
    y = deaths_by_date
)
plt.xticks(rotation=45)
plt.xlabel('Time')
plt.ylabel('Deaths')
plt.title("US Coronavirus deaths from Jan 2020 to Apr 2020")


plt.show()
```

The plot above is not especially useful, partially because we only have data halfway to April, partially because we're generalizing this data for the whole US. But we also have to recognize that our data looks exponential, and plotting our data versus time may not give us the best idea of what's going on.
